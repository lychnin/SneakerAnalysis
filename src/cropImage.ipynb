```python
import cv2

# 创建一个人脸检测器
detector = cv2.FaceDetectorYN.create("yunet.onnx")

# 读取图像
img = cv2.imread()

# 检测人脸
faces = detector.detect(img)

# 遍历每个人脸
for face in faces:
    # 获取人脸位置
    x, y, w, h = face.getRect()
    # 绘制矩形框
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# 创建一个人脸识别器
recognizer = cv2.FaceRecognizerSF.create("sfnet.onnx")

# 加载已知的人脸数据
recognizer.addLabel("Alice", "alice.jpg")
recognizer.addLabel("Bob", "bob.jpg")
recognizer.addLabel("Charlie", "charlie.jpg")

# 遍历每个人脸
for face in faces:
    # 获取人脸位置
    x, y, w, h = face.getRect()
    # 获取人脸特征向量
    feature = face.getFeature()
    # 识别人脸身份
    label, score = recognizer.recognize(feature)
    # 绘制文本框
    cv2.putText(img, label + " (" + str(score) + ")", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

# 显示图像
cv2.imshow("Face Recognition", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
