```python
import pandas as pd
import re
import numpy 
import matplotlib
import jieba
```

```python
data=pd.read_csv(r"D:\VSCode\Project\crawler\Sneaker\intermediateResult\NOWREdemo.csv")
```

```python
title=[x for x in data["title"]]
source=[x for x in data["source"]]
```

```python
ins=[]
brand=[]
for x in source:
    pattern="Instagram.+"
    if re.findall(pattern,x):
        ins.append(re.findall(pattern,x))
    else:
        brand.append(x)
```

```python
brand=set(brand)
brand=list(brand)
```

```python
product=[]
for x in title:
    a=re.findall("[a-zA-Z0-9]+\w+|[a-zA-Z0-9]+\s",x)
    if a:
        a=[x.strip() for x in a]
        odd=re.findall("\w+",x)
        # 将表达联名关系的"与"替换为"x"
        if "与" in odd:
            loc=odd.index("与")
            odd.insert(loc,"x")
            odd.pop(loc+1)
            x=" ".join(odd)
            a=re.findall("[a-zA-Z0-9]+\w+|[a-zA-Z0-9]+\s",x)
            a=[x.strip() for x in a]
        if a[0]=="x":
            frontx=re.findall("\w+",x)
            index=frontx.index("x")
            # "x"之前的中文单词
            patch=frontx[index-1]
            a.insert(0,patch)
        product.append(" ".join(a))
        #print(" ".join(a))
    else:
        product.append("Extract error")
```

```python
len(product)
```

```python
ProductPd=pd.DataFrame()
ProductPd["product"]=product
ProductPd["source"]=data["source"]
ProductPd["image"]=data["image"]
ProductPd["date"]=data["date"]
```

```python
images=[]
for imageEle in ProductPd["image"]:
    if imageEle =="[]":
        images.append("None")
    else:
        # 去掉"[]"
        medium=imageEle.strip("[]")
        # 将几个链接分开
        all=medium.split(",")
        ima=[]
        for x in all:
            # 去掉左侧的" '"
            x=x.strip(" '")
            # 去掉右侧的"'"
            x=x.strip("'")
            ima.append(x)
        images.append(ima)
```

```python
import requests
from requests.exceptions import ProxyError
from requests.exceptions import ConnectionError
from requests.exceptions import SSLError
from requests.exceptions import InvalidSchema
import time
#图片链接
for i in range(14184,len(images)):
    if images[i]== 'None':
        continue
    else:
        num=0
        for im in images[i]:    
            if re.search("http",im):
                num+=1
                print(str(i)+"-"+str(num))
                image_url= im
                headers ={
                'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'
                }
                try:
                    r = requests.get(image_url,headers=headers)
                except ProxyError as e:
                    print("ProxyError")
                    time.sleep(2)
                    r = requests.get(image_url,headers=headers)
                except ConnectionError as E:
                    print("CinnectionError")
                    time.sleep(500)
                    r = requests.get(image_url,headers=headers)
                except SSLError as E2:
                    print("CinnectionError")
                    time.sleep(500)
                    r = requests.get(image_url,headers=headers)
                except InvalidSchema as ee:
                    print("InvalidSchema")
                    continue
                # 下载图片
                with open(f"d:/VSCode/Project/crawler/Sneaker/image/{product[i]}-{num}.jpg" ,mode = "wb") as f:
                    f.write(r.content) #图片内容写入文件
                    f.close()
                    print(i)
```
